```julia
iris = load_iris();
selectrows(iris,1:3) |> pretty 
```

```julia
schema(iris)
```

```julia
import DataFrames
iris_data = DataFrames.DataFrame(iris)
y,X = unpack(iris_data, ==(:target); rng=123)
first(X,3) |> pretty 
```

```julia
using StatsPlots
@df iris plot(:sepal_length, :sepal_width, group=:target)
```

```julia
first(iris_data,5) |> pretty 
```

```julia
models(matching(X,y))
```

```julia
try
    Tree = @load DecisionTreeClassifier pkg=DecisionTree
catch
    Pkg.add("MLJDecisionTreeInterface")
end
Tree = @load DecisionTreeClassifier pkg=DecisionTree
```

```julia
tree = Tree()
```

```julia
evaluate(tree,X,y, resampling = CV(shuffle=true), measures = [log_loss,accuracy],verbosity=1)
```

```julia
mach_ = machine(tree,X,y)

train,test = partition(eachindex(y),0.7)
fit!(mach_,rows=train)
```

```julia
y_pred = predict(mach_,X[test,:])
```

```julia
evaluate!(mach_,resampling = Holdout(fraction_train=0.7), measure = [accuracy],verbosity=1)
```
